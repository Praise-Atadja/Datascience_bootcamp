```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#create sample data for IGTV
data = {
    'Date': pd.date_range(start='2023-01-01', periods=12, freq='M'),
    'Views': [100000, 110000, 120000, 130000, 140000, 150000, 160000, 170000, 180000, 190000, 200000, 210000],
    'Average_Watch_Time': [5.2, 5.4, 5.5, 5.7, 5.9, 6.0, 6.2, 6.3, 6.5, 6.7, 6.8, 7.0],
    'Audience_Retention': [50, 55, 60, 62, 65, 67, 70, 72, 75, 78, 80, 82],
    'Engagement_Rate': [3.2, 3.5, 3.7, 4.0, 4.2, 4.5, 4.7, 5.0, 5.2, 5.5, 5.7, 6.0],
    'Follower_Growth': [1000, 1100, 1200, 1300, 1400, 1500, 1600, 1700, 1800, 1900, 2000, 2100],
    'Revenue': [5000, 5500, 6000, 6200, 6500, 6700, 7000, 7200, 7500, 7800, 8000, 8200]
}
```

```python
#create a dataframe from the data
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(12, 6))
plt.title("IGTV Metrics Over Time")
plt.plot(df['Date'], df['Views'], label='Views', marker='o')
plt.plot(df['Date'], df['Average_Watch_Time'], label='Avg. Watch Time', marker='o')
plt.plot(df['Date'], df['Audience_Retention'], label='Audience Retention', marker='o')
plt.plot(df['Date'], df['Engagement_Rate'], label='Engagement Rate', marker='o')
plt.plot(df['Date'], df['Follower_Growth'], label='Follower Growth', marker='o')
plt.plot(df['Date'], df['Revenue'], label='Revenue', marker='o')
plt.xlabel('Date')
plt.ylabel('Metrics')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
